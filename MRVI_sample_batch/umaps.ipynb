```python
import anndata as ad
import pandas as pd
import scanpy as sc
import scvi
```

```python
adata = ad.read_h5ad("/home/albert.baichorov/ImSysAging/data/scRNA-seq_MRVI_sample_batch.h5ad")
print(adata)
```

```python
from matplotlib import rcParams

FIGSIZE = (10, 10)
rcParams["figure.figsize"] = FIGSIZE
```

# Z embedings

```python
sc.pp.neighbors(adata, use_rep="Z_mrvi")
sc.tl.umap(adata, min_dist=0.3)

sc.pl.umap(adata, color=["primary_type"], size=5)
sc.pl.umap(adata, color=["sampleName"], size=10)
```
